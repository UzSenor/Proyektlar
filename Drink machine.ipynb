```python
Baza = {
    "qoldiq" : {
        "pepsi" : 3,
        "cola"  : 5,
        "fanta" : 4,
        "sprite": 7
    },
    "narx" : {
        "pepsi" : 2100,
        "cola"  : 2200,
        "fanta" : 1800,
        "sprite": 1900
    },
    "pul_birlik" : {
        "100"   : 5,
        "200"   : 8,
        "500"   : 7,
        "1000"  : 9,
        "2000"  : 7,
        "5000"  : 5,
        "10000" : 4,
        "20000" : 4,
        "50000" : 3,
        "100000": 2
    },
    "summa" : 518600
}

# Funksiyalar
def menu(baza):
    print("Pepsi:",Baza["qoldiq"]["pepsi"],"ta",
          "\nCola:",Baza["qoldiq"]["cola"], "ta",
          "\nFanta:",Baza["qoldiq"]["fanta"], "ta",
          "\nSprite:",Baza["qoldiq"]["sprite"], "ta")

def narx(baza):
    print("Pepsi:",Baza["narx"]["pepsi"],"so'm",
          "\nCola:",Baza["narx"]["cola"], "so'm",
          "\nFanta:",Baza["narx"]["fanta"], "so'm",
          "\nSprite:",Baza["narx"]["sprite"], "so'm")
    
def pul(baza):
    print("100 so'mlik:",Baza["pul_birlik"]["100"],"ta",
          "\n200 so'mlik:",Baza["pul_birlik"]["200"], "ta",
          "\n500 so'mlik:",Baza["pul_birlik"]["500"], "ta",
          "\n1 000 so'mlik:",Baza["pul_birlik"]["1000"], "ta",
          "\n2 000 so'mlik:",Baza["pul_birlik"]["2000"], "ta",
          "\n5 000 so'mlik:",Baza["pul_birlik"]["5000"], "ta",
          "\n10 000 so'mlik:",Baza["pul_birlik"]["10000"], "ta",
          "\n20 000 so'mlik:",Baza["pul_birlik"]["20000"], "ta",
          "\n50 000 so'mlik:",Baza["pul_birlik"]["50000"], "ta",
          "\n100 000 so'mlik:",Baza["pul_birlik"]["100000"], "ta",)
    
def summa(baza):
    print("Qoldiq summa:", Baza["summa"], "so'm")
    
def sotib_olish(baza):
    x = True
    while x:
        # Ichimlik tanlash
        ichimlik = input("Ichimlik turini yozing(pepsi, cola, fanta, sprite, off):\n")
        print("Bazada", Baza["qoldiq"][ichimlik] ,"ta mahsulot bor")
        summ = 0
        for i in Baza["qoldiq"].keys():
            if i == ichimlik:
                # Ichimlik sonini kiritish
                soni = int(input("Ichimlik sonini kiriting:\n"))
                # Ichimlik sonini tekshiradi
                if soni <= Baza["qoldiq"][i]:
                    narxi = soni * Baza["narx"][i]
                    print(narxi)
                    # Pul to'lash
                    print("Pul soling:")
                    a = int(input("100 so'mlik: "))
                    b = int(input("200 so'mlik: "))
                    c = int(input("500 so'mlik: "))
                    d = int(input("1 000 so'mlik: "))
                    e = int(input("2 000 so'mlik: "))
                    f = int(input("5 000 so'mlik: "))
                    g = int(input("10 000 so'mlik: "))
                    h = int(input("20 000 so'mlik: "))
                    i = int(input("50 000 so'mlik: "))
                    j = int(input("100 000 so'mlik: "))
                    # To'langan pulni bazadagi mos birliklarga qo'shadi
                    Baza["pul_birlik"]["100"] += a
                    Baza["pul_birlik"]["200"] += b
                    Baza["pul_birlik"]["500"] += c
                    Baza["pul_birlik"]["1000"] += d
                    Baza["pul_birlik"]["2000"] += e
                    Baza["pul_birlik"]["5000"] += f
                    Baza["pul_birlik"]["10000"] += g
                    Baza["pul_birlik"]["20000"] += h
                    Baza["pul_birlik"]["50000"] += i
                    Baza["pul_birlik"]["100000"] += j
                    
                    summ = summ + a*100 + b*200 + c*500 + d*1000 + e*2000 + f*5000 + g*10000 + h*20000 + i*50000 + j*100000
                    # Umumiy summani qo'shadi
                    Baza["summa"] += summ
                    print(summ)
                    # To'langan pul bilan mahsulot narhini solishtiradi
                    if narxi < summ:
                        Baza["qoldiq"][ichimlik] -= soni
                        print("Yoqimli ishtaha!!!")
                        qoldi = summ - narxi
                        print(f"Sizda {qoldi} so'm qoldi. Yana mahsulot olasizmi yoki qaytib beraymi? h/y")
                    elif narxi == summ:
                        Baza["qoldiq"][ichimlik] -= soni
                        
                        print("Yoqimli ishtaha!!!")
                        x = False
                    # Pul yetmay qolsa qaytadan pul so'raydi
                    else:
                        print("Siz bergan pulingiz yetmaydi. Yana qo'shing!")
                        a = int(input("100 so'mlik: "))
                        b = int(input("200 so'mlik: "))
                        c = int(input("500 so'mlik: "))
                        d = int(input("1 000 so'mlik: "))
                        e = int(input("2 000 so'mlik: "))
                        f = int(input("5 000 so'mlik: "))
                        g = int(input("10 000 so'mlik: "))
                        h = int(input("20 000 so'mlik: "))
                        i = int(input("50 000 so'mlik: "))
                        j = int(input("100 000 so'mlik: "))
                        summ = summ + a*100 + b*200 + c*500 + d*1000 + e*2000 + f*5000 + g*10000 + h*20000 + i*50000 + j*100000
                        print(summ)
                # Ichimlikning bazadagi soni yetmasa
                else:
                    print("Bazada bu mahsulot yetmaydi. Kamroq tanlang yoki boshqa mahsulot tanlang!")
                    
                break
        # Ichimlik nomini tekshiradi
        else:
            print("Bizda bunday ichimlik yo'q!")
            if ichimlik == "off":
                x = False
    

# Asosiy qism   
x = True
while x:
    amal = input("Siz nima qilmoqchisiz?(menu, summa, narx, pul, sotib olish, off):\n")
    
    if amal == "menu":
        menu(amal)
        
    elif amal == "narx":
        narx(amal)
        
    elif amal == "pul":
        pul(amal)
        
    elif amal == "summa":
        summa(amal)
        
    elif amal == "sotib olish":
        sotib_olish(amal)
        
    elif amal == "off":
        x = False
```
